```python
# with matplotlib
import matplotlib.pyplot as plt

# Provided output data
gender_counts = {
    'male': 62,
    'female': 19,
    'no gender': 3,
    'none': 2,
    'hermaphrodite': 1
}

# Plotting the distribution of Star Wars characters by gender
plt.bar(gender_counts.keys(), gender_counts.values())
for i, count in enumerate(gender):
    plt.text(i, count, str(count))
plt.title('Gender Distribution')
plt.show()
```

```python
# with pandas
import matplotlib.pyplot as plt

# Assuming gender is your pandas Series containing the gender distribution

fig, ax = plt.subplots()
ax.bar(gender.index, gender)

ax.set_title('Gender Distribution', fontdict={'fontsize': 16, 'color': 'r'})  # Corrected argument name
for i, count in enumerate(gender):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=8)

plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

## <p style="background-color:#9d4f8c; font-family:newtimeroman; color:#FFF9ED; font-size:150%; text-align:center; border-radius:10px 10px;">TASK_3 - Plotting the MASS unde 3 groups: "High_mass", "Normal_mass", "Low_mass" </p>

```python
# Define the bins and labels for categorizing mass
bins = [-float('inf'), 50, 100, float('inf')]
labels = ['Low_mass', 'Normal_mass', 'High_mass']

# Categorize the mass into three groups
df['mass_cat'] = pd.cut(df['mass'], bins=bins, labels=labels)

# Count the number of characters in each category, including NaN
mass_counts = df['mass_cat'].value_counts(dropna=False)

# Display the counts
print(mass_counts)
```

```python

# Assuming df is your DataFrame containing the information

# Define the bins and labels for categorizing mass
bins = [-float('inf'), 50, 100, float('inf')]
labels = ['Low_mass', 'Normal_mass', 'High_mass']

# Categorize the mass into three groups
df['mass_cat'] = pd.cut(df['mass'], bins=bins, labels=labels)

# Count the number of characters in each group, excluding NaN
mass_counts = df['mass_cat'].value_counts(dropna=True)

# Plotting the mass distribution into three groups
mass_counts.plot(kind='bar', rot=90)  # Rotate x-axis labels
# Annotate the bars with their respective counts
for i, count in enumerate(mass_counts):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=8)
plt.xlabel('mass_cat')
plt.show()
```

```python
# with matplotlib
import matplotlib.pyplot as plt

# Assuming df is your DataFrame containing the information
# Define the bins and labels for categorizing mass
bins = [-float('inf'), 50, 100, float('inf')]
labels = ['Low_mass', 'Normal_mass', 'High_mass']

# Categorize the mass into three groups
df['mass_cat'] = pd.cut(df['mass'], bins=bins, labels=labels)

# Count the number of characters in each group, excluding NaN
mass_counts = df['mass_cat'].value_counts(dropna=True)
# Annotate the bars with their respective counts
for i, count in enumerate(mass_counts):
    plt.text(i, count, str(count), ha='center', va='bottom', fontsize=8)
# Plotting the mass distribution into three groups using matplotlib
plt.bar(mass_counts.index, mass_counts.values)  # Change color if desired
plt.show()
```

```python
# with pandas
import pandas as pd
import matplotlib.pyplot as plt
# Assuming df is your DataFrame containing the information

# Count the occurrences of each species
species_counts = df['species'].value_counts()

# Select the top 5 species
top_5_species = species_counts.head(5)

# Plotting the top 5 species using pandas
ax = top_5_species.plot(kind='bar', figsize=(10, 6), xlabel='species', ylabel='Frequency')
ax.set_xticklabels(top_5_species.index, rotation=90)  # Rotate x-axis labels for better readability

# Add labels to the bars
for i, v in enumerate(top_5_species.values):
    ax.text(i, v, str(v), ha='center', va='bottom', fontsize=8)

# Set the title of the plot with blue color
ax.set_title('Species', color='blue')

plt.show()

```
