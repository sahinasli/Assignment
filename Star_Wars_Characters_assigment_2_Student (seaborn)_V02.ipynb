```python
#describe
# Define a function to calculate the frequency
def describe_data(df):
    """
    Describe the DataFrame by providing count, unique values, top value, and frequency for each column.

    Parameters:
    df (DataFrame): The pandas DataFrame containing the data.

    Returns:
    DataFrame: A DataFrame summarizing the count, unique values, top value, and its frequency for each column.
    """
    description = df.describe(include='all').transpose()  # Get basic statistics for each column
    unique_counts = df.nunique()  # Get unique value counts for each column

    # Add additional information to the description DataFrame
    description['count'] = description['count'].astype(int)
    description['unique'] = unique_counts
    description['top'] = df.mode(dropna=True).iloc[0]
    description['freq'] = df.apply(lambda x: x.value_counts().idxmax())

    return description

# Example usage:
# Assuming 'df' is your DataFrame containing the Star Wars character data
description = describe_data(df)
print(description)
```

```python
# with seaborn Swarmplot


# Categorize mass into light, medium, and heavy
def categorize_mass(mass):
    if mass is None:
        return 'Unknown'
    elif mass < 50:
        return 'Low_mass'
    elif mass < 100:
        return 'Normal_mass'
    else:
        return 'High_mass'

df['mass_cat'] = df['mass'].apply(categorize_mass)

# Plot Swarmplot
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.swarmplot(x="gender", y="height", hue="mass_cat", data=df)
plt.ylabel('height')
plt.legend(title='mass_cat')
plt.show()
```
